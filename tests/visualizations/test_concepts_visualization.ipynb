```python
import sys
import torch
from matplotlib import pyplot as plt

sys.path.append("../..")
from interpreto.visualizations.concepts.concepts_highlight import ConceptHighlightVisualization
```

```python
# Concepts: 9 concepts (with inputs attributions for each output word)
inputs_sentences = [
    ["A", "B", "C", "one", "two", "three"],
    ["This", " ", "is an", " example of ", "a longer", " ", "sen", "ten", "ce", " ", "with ", "more", " ", "word","s", " and ", "spaces", " ", "bet", "ween", " ", "word", "s"],
]
# outputs_sentence = ["do", "re", "mi"]
outputs_sentence = ["A", "B", "C", " ", "look", " ", "behind", " ", "you", " : ", "there", " ", "is", " ", "a", " ", "monster", " ", "under", " ", "the", " ", "bed", "!!!"]
nb_concepts = 9

inputs_attributions = [torch.rand((len(outputs_sentence), len(inputs_sentence), nb_concepts)) for inputs_sentence in inputs_sentences]
outputs_attributions = torch.rand((len(outputs_sentence), len(outputs_sentence), nb_concepts))
```

```python
colors_set1 = plt.get_cmap("Set1").colors
```

```python
viz = ConceptHighlightVisualization(
    inputs_sentences=inputs_sentences,
    inputs_attributions=inputs_attributions,
    outputs_words=outputs_sentence,
    outputs_attributions=outputs_attributions,
    concepts_colors=colors_set1,
)
```
